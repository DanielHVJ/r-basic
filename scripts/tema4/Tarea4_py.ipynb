```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 256

X = np.linspace(-np.pi, np.pi, n, endpoint=True)

Y = np.sin(2 * X) 

plt.plot(X, Y + 1, color='blue', alpha=1.00)

plt.plot(X, Y - 1, color='red', alpha=1.00)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

ax1.fill_between(X, 0, Y)
ax1.set_ylabel('First plot')

ax2.fill_between(X, 0, Y, where = Y <= 0, facecolor='pink', interpolate=False)
ax2.fill_between(X, 0, Y, where = Y >= 0, facecolor='blue', interpolate=False)
ax2.set_ylabel('Second plot')
```
